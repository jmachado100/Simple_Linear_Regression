# Simple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = np.array([1.1,1.2,1.3,1.5,2,2.2,2.5,2.9,3,3.2,
     3.2,3.5,3.7,3.8,3.9,4,4,4.1,4.3,4.5,
     4.7,4.9,5.1,5.3,5.5,5.9,6,6.2,6.5,6.8,
     7.1,7.9,8.2,8.5,8.7,9,9.5,9.6,10.3,10.5]).reshape(-1,1)

y = np.array([39,42,46,37,43,39,48,56,60,54,
    64,60,57,60,63,55,56,57,59,61,
    64,67,66,83,82,81,93,91,90,91,
    98,101,113,111,109,105,116,112,122,121]).reshape(-1,1)
```

```python
df = pd.DataFrame({'X': X.flatten(), 'y': y.flatten()})
```

```python
plt.scatter(df['X'], df['y'])
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[['X']], df['y'], train_size=0.7, test_size=0.3, random_state=12)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print("Slope: ", model.coef_[0])
print("Intercept: ", model.intercept_)
```

```python
y_predict = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_predict)
print("Mean Squared Error: ", mse)
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_predict, color='red', linewidth=2)
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
